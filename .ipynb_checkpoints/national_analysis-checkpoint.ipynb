## All necessary imports

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import cross_validate, cross_val_score, GridSearchCV, KFold, train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from itertools import combinations
from sklearn.linear_model import MultiTaskLassoCV
from xgboost import XGBRegressor
from sklearn.decomposition import PCA
import smogn
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

## Data Input and Pre-processing

```python
data_path = 'data/'
df = pd.read_csv(data_path + 'historic_policy_data.csv', index_col="Year")

predictors = ["% Suspension", "% Year Removal", "% Transfer", "% security staff", "% law enforcement with firearm", \
        "% metal detectors", "% security cameras", "% locked doors", "% locker checks", "Minority 0-25% security staff", \
        "Minority 25-50% security staff", "Minority 50-75% security staff", "Minority 75-100% security staff", \
        "Minority 0-25% law enforcement with firearm", "Minority 25-50% law enforcement with firearm", "Minority 50-75% law enforcement with firearm", \
        "Minority 75-100% law enforcement with firearm" ]
responses = ["% schools violent incident", "Violent crime per 1000"]

df_first = df.loc[:, predictors + responses]
df_first
```

```python
scaler = StandardScaler()
df_first[predictors] = scaler.fit_transform(df_first[predictors])
normalizer = MinMaxScaler()
df_first[predictors] = normalizer.fit_transform(df_first[predictors])
df_first
```

```python
scaler = StandardScaler()
df_first[responses] = scaler.fit_transform(df_first[responses])
normalizer = MinMaxScaler()
df_first[responses] = normalizer.fit_transform(df_first[responses])
df_first
```

### Correlation Testing & Linear Regression

```python
g = sns.pairplot(df, x_vars=predictors, y_vars=responses, kind='reg', height=3, aspect=0.6)
axis_bounds = {"% schools violent incident" : [0, 100], "Violent crime per 1000" : [0, 50]}

for ax in g.axes.flatten():
    ylab = ax.get_ylabel()
    if len(ylab) == 0: 
        continue
    ax.set_ylim(axis_bounds[ylab])
```

```python
coefficients_of_variation = []
for feature in df:
    curr_cv_val = np.std(df[feature]) / np.mean(df[feature])
    coefficients_of_variation.append(curr_cv_val)
df_coefficients = pd.DataFrame({"Feature" : df.columns, "Coefficient of Variation" : coefficients_of_variation})
print("Coefficient of Variation for All Features:\n")
print(df_coefficients, "\n")
plt.matshow(df.corr())
plt.colorbar()
```

```python
linreg_model = LinearRegression()
for predictor in predictors:
    training = sm.add_constant(df_first[[predictor]])
    for response in responses:
        results = sm.OLS(df_first[response], training).fit()
        print("\nPredictor:", predictor, "\nResponse:", response + "\n", results.summary())
```

We can see in the above results that there exists statistically significant associations between some predictors and the response, where our OLS regression outputs p-values under 0.05 (our threshold for explaining associations by chance).

```python
pairs = [['% Suspension', '% security staff'], ['% Suspension', '% law enforcement with firearm'], ['% Suspension', '% metal detectors'], \
         ['% Suspension', '% security cameras'], ['% Suspension', '% locked doors'], ['% Suspension', '% locker checks'], \
         ['% Year Removal', '% security staff'], ['% Year Removal', '% law enforcement with firearm'], \
         ['% Year Removal', '% metal detectors'], ['% Year Removal', '% security cameras'], ['% Year Removal', '% locked doors'], \
         ['% Year Removal', '% locker checks'], ['% Transfer', '% security staff'], ['% Transfer', '% law enforcement with firearm'], \
         ['% Transfer', '% metal detectors'], ['% Transfer', '% security cameras'], ['% Transfer', '% locked doors'], ['% Transfer', '% locker checks'], \
         ['% security staff', '% law enforcement with firearm'], ['% security staff', '% metal detectors'], ['% security staff', '% security cameras'], ['% security staff', '% locked doors'], \
         ['% security staff', '% locker checks'], ['% law enforcement with firearm', '% metal detectors'], ['% law enforcement with firearm', '% security cameras'], ['% law enforcement with firearm', '% locked doors'], \
         ['% law enforcement with firearm', '% locker checks'], ['% metal detectors', '% security cameras'], ['% metal detectors', '% locked doors'], ['% metal detectors', '% locker checks'], \
         ['% security cameras', '% locked doors'], ['% security cameras', '% locker checks'], ['% locked doors', '% locker checks']]
for pair in pairs:
    pair_string = pair[0] + " * " + pair[1]
    pair_values = df_first[pair].values
    pairwise_interaction = pd.DataFrame(pair_values[:, 0] * pair_values[:, 1], columns=[pair_string])
    for response in responses:
        training = sm.add_constant(pairwise_interaction)
        results = sm.OLS(df_first[response].values, training).fit()
        print("\nPredictor:", pair_string, "\nResponse:", response + "\n", results.summary())
```

```python
dataFrames = []
for predictor in predictors:
    dataFrames.append(pd.DataFrame(df_first[predictor].values, columns=[predictor]))
for pair in pairs:
    pair_values = df_first[pair].values
    pairwise_interaction = pd.DataFrame(pair_values[:, 0] * pair_values[:, 1], columns=[pair[0] + " * " + pair[1]])
    dataFrames.append(pairwise_interaction)
training = pd.concat(dataFrames, axis=1)
training = sm.add_constant(training)
for response in responses:
    results = sm.OLS(df_first[response].values, training).fit()
    print("\nMultiple Regression Model\n", " Response:", response + "\n", results.summary())
```

### K-Neighbors Regression

```python
k_values = list(range(1, 6))
training_error = []
for curr_k in k_values:
    print("k value:", curr_k)
    classifier_output = KNeighborsRegressor(n_neighbors=curr_k)
    classifier_output.fit(df_first[predictors], df_first[responses])
    curr_error = mean_squared_error(df_first[responses], classifier_output.predict(df_first[predictors]))
    curr_k += 1
    print("Ground truth:\n", df_first[responses], "\n\nPredictions:\n", classifier_output.predict(df_first[predictors]))
    print("Training error:", curr_error, "\n")
```

```python
predictors_subset = ["% Suspension", "% Year Removal", "% Transfer", "% security staff", "% law enforcement with firearm", \
        "% metal detectors", "% security cameras", "% locked doors", "% locker checks"]
k_values = list(range(1, 6))
training_error = []
for curr_k in k_values:
    print("k value:", curr_k)
    classifier_output = KNeighborsRegressor(n_neighbors=curr_k)
    classifier_output.fit(df_first[predictors_subset], df_first[responses])
    curr_error = mean_squared_error(df_first[responses], classifier_output.predict(df_first[predictors_subset]))
    curr_k += 1
    print("Ground truth:\n", df_first[responses], "\n\nPredictions:\n", classifier_output.predict(df_first[predictors_subset]))
    print("Training error:", curr_error, "\n")
```

### LASSO

```python
lasso_model = MultiTaskLassoCV(alphas=[0.00001, 0.00005, 0.0001, 0.005, 0.001, 0.05, 0.01, 0.1, 0.5, 1, 10, 100, 1000], random_state=0)
lasso_model = lasso_model.fit(df_first[predictors], df_first[responses])
print("Best alpha for LASSO Model:", lasso_model.alpha_)
print("LASSO Model Testing Error:", mean_squared_error(df_first[responses], lasso_model.predict(df_first[predictors])))
selected_variables_percent = pd.DataFrame({"Variable" : predictors, "Coefficient_%schools_violent_incident" : lasso_model.coef_[0]})
selected_variables_per1000 = pd.DataFrame({"Variable" : predictors, "Coefficient_violent_crime_per1000" : lasso_model.coef_[1]})
selected_variables_percent = selected_variables_percent.loc[selected_variables_percent["Coefficient_%schools_violent_incident"] != 0]
selected_variables_per1000 = selected_variables_per1000.loc[selected_variables_per1000["Coefficient_violent_crime_per1000"] != 0]
selected_variables = selected_variables_percent.merge(selected_variables_per1000, left_on="Variable", right_on="Variable")
print("\nVariables Selected:\n")
print(selected_variables)
```

### PCA

```python
scores = []
component_values = [i for i in range(1, 7)]
for num_components in component_values:
    pca = PCA(n_components=num_components, random_state=0)
    training = pca.fit_transform(df_first[predictors])
    linreg_model = LinearRegression()
    linreg_model.fit(training, df_first[responses])
    curr_score = mean_squared_error(df_first[responses], linreg_model.predict(training))
    scores.append(curr_score)

component_summary = pd.DataFrame({ "n_components Value" : component_values, "Mean Squared Error" : scores })
best_n_component = component_summary.sort_values(by="Mean Squared Error")
print(best_n_component)
```

### XGBRegressor

```python
scores = []
alpha_values = [0.0000001, 0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 5, 10, 100, 1000]
for alpha in alpha_values:
    xgb_model = XGBRegressor(reg_alpha=alpha, reg_lambda=0, random_state=0)
    xgb_model.fit(df_first[predictors], df_first[responses])
    curr_score = mean_squared_error(df_first[responses], xgb_model.predict(df_first[predictors]))
    scores.append(curr_score)

alpha_summary = pd.DataFrame({ "L1 Regression alpha Value" : alpha_values, "Mean Squared Error" : scores })
best_alpha_value = alpha_summary.sort_values(by="Mean Squared Error")
print(best_alpha_value)
```
