## Data Input and Pre-processing

```python
data_path = 'data/'
df = pd.read_csv(data_path + 'historic_policy_data.csv', index_col="Year")

predictors = ["% Suspension", "% Year Removal", "% Transfer", "% security staff", "% law enforcement with firearm", \
        "% metal detectors", "% security cameras", "% locked doors", "% locker checks"]
responses = ["% schools violent incident", "Violent crime per 1000"]

df_first = df.loc[:, predictors + responses]
df_first
```

```python
scaler = StandardScaler()
df_first[predictors] = scaler.fit_transform(df_first[predictors])
normalizer = MinMaxScaler()
df_first[predictors] = normalizer.fit_transform(df_first[predictors])
df_first
```

```python
scaler = StandardScaler()
df_first[responses] = scaler.fit_transform(df_first[responses])
normalizer = MinMaxScaler()
df_first[responses] = normalizer.fit_transform(df_first[responses])
df_first
```

### Correlation Testing & Linear Regression

```python
g = sns.pairplot(df, x_vars=predictors, y_vars=responses, kind='reg', height=3, aspect=0.6)
axis_bounds = {"% schools violent incident" : [0, 100], "Violent crime per 1000" : [0, 50]}

for ax in g.axes.flatten():
    ylab = ax.get_ylabel()
    if len(ylab) == 0: 
        continue
    ax.set_ylim(axis_bounds[ylab])
```

```python
linreg_model = LinearRegression()
for response in responses:
    for predictor in predictors:
        model_fit = linreg_model.fit(df_first[[predictor]], df_first[response])
        training = sm.add_constant(df_first[[predictor]])
        results = sm.OLS(df_first[response], training).fit()
        print("\nPredictor:", predictor, "\nResponse:", response + "\n", results.summary())
```

We can see in the above results that there exists statistically significant associations between some predictors and the response, where our OLS regression outputs p-values over 0.05.
